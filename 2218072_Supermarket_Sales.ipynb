# **3: Membaca dan Melihat Data**

```python
df = pd.read_csv("supermarket_sales.csv")

# Melihat 5 baris pertama
print("=== Data Awal ===")
display(df.head())

# Melihat informasi tipe data dan missing values
print("\n=== Info Dataset ===")
df.info()

# Melihat ringkasan statistik untuk data numerik
print("\n=== Statistik Awal ===")
display(df.describe())
```

# **4: Menghapus Data Duplikat**

```python
print(f"Jumlah data duplikat sebelum dibersihkan: {df.duplicated().sum()}")
df.drop_duplicates(inplace=True)
print(f"Jumlah data duplikat setelah dibersihkan: {df.duplicated().sum()}")
```

# **5: Menangani Missing Values**

```python
# Mengecek jumlah missing value
print("\nJumlah missing values per kolom:")
print(df.isnull().sum())

# Jika ada, isi missing value:
for col in df.columns:
    if df[col].dtype == 'object':  # Jika kolom teks
        df[col].fillna('Unknown', inplace=True)
    else:  # Jika kolom numerik
        df[col].fillna(df[col].median(), inplace=True)

print("\nJumlah missing values setelah penanganan:")
print(df.isnull().sum())
```

# **6: Membersihkan Format Kolom**

```python
# Format tanggal
if 'Date' in df.columns:
    df['Date'] = pd.to_datetime(df['Date'], errors='coerce')

# Standarisasi teks di kolom 'City'
if 'City' in df.columns:
    df['City'] = df['City'].str.strip().str.title()
```

# **7: Menghapus Outlier**

```python
if 'Total' in df.columns:
    Q1 = df['Total'].quantile(0.25)
    Q3 = df['Total'].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    before_rows = df.shape[0]
    df = df[(df['Total'] >= lower_bound) & (df['Total'] <= upper_bound)]
    after_rows = df.shape[0]

    print(f"\nJumlah data outlier yang dihapus: {before_rows - after_rows}")
```

# **8: Simpan Data Bersih**

```python
df.to_csv("supermarket_sales_clean.csv", index=False)
print("Data bersih berhasil disimpan sebagai 'supermarket_sales_clean.csv'")
files.download("supermarket_sales_clean.csv")
```

# 10: Data Enrichment

```python
# Pastikan kolom Date sudah dalam format datetime
if 'Date' in df.columns:
    df['DayOfWeek'] = df['Date'].dt.day_name()    # Nama hari, contoh: Monday
    df['Month'] = df['Date'].dt.month_name()      # Nama bulan, contoh: January

# Membuat kategori penjualan
if 'Total' in df.columns:
    bins = [0, 100, 250, np.inf]
    labels = ['Low', 'Medium', 'High']
    df['SalesCategory'] = pd.cut(df['Total'], bins=bins, labels=labels)

print("\n=== Data Setelah Enrichment ===")
display(df.head())
```
